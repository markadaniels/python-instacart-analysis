```python
# Retrieve departments data set to find the meaning behind the value 4 in the "department_id" column
df_dep = pd.read_csv(os.path.join(path, '03 Scripts', 'Original Data', 'departments.csv'), index_col = False)
```

#### Transposing departments.csv Data

```python
# Explore the data set and we see just one row and 22 columns
df_dep.head()
```

```python
# Create a new dataframe from the transposed version
df_dep_t = df_dep.T
```

```python
# Review the new dataframe and we see our column headers in the second row instead of the first row
df_dep_t
```

```python
# Pythons numbering starts at zero, so we can add an index column
# This will have the headers in row zero and the data starts from 1
df_dep_t.reset_index()
```

```python
# Create a new header variable that will hold the headers that are in row zero
new_header = df_dep_t.iloc[0]
```

```python
# Create a new dataframe to hold the rest of the data starting from row 1
df_dep_t_new = df_dep_t[1:]
```

```python
# Assign new_header to df_dep_t_new as column headers
df_dep_t_new.columns = new_header
```

```python
# Review df_dep_t_new
df_dep_t_new
```

#### Create data dictionary

```python
# Convert the new dataframe "df_dep_t_new" to a data dictionary
data_dict = df_dep_t_new.to_dict('index')
```
